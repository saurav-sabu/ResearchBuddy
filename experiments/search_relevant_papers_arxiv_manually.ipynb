```python
import requests
import feedparser
```

```python
def search_arxiv_papers(query,limit=5):

    base_url = "http://export.arxiv.org/api/query?"
    search_url = f"{base_url}search_query=all:{query}&start=0&max_results={limit}"

    response = requests.get(url=search_url)
    feed = feedparser.parse(response.text)

    papers = []

    for i, entry in enumerate(feed.entries,1):
        title = entry.title
        authors = [author.name for author in entry.authors]
        published = entry.published
        summary = entry.summary
        pdf_url = None

        for link in entry.links:
            if link.type == "application/pdf":
                pdf_url = link.href

        papers.append(
            {
                "title":title,
                "authors":authors,
                "published":published,
                "summary":summary,
                "pdf_links":pdf_url
            }
        )

        print(f"Title: {title}")
        print(f"Authors: {",".join(authors)}")
        print(f"Published: {published}")
        print(f"PDF URL: {pdf_url}")

    return papers

```

```python
arxiv_results = search_arxiv_papers("machine learning",limit=3)
```
